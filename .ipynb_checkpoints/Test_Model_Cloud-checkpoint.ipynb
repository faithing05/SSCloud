```python
import torch

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –¥–æ—Å—Ç—É–ø–Ω–æ—Å—Ç—å CUDA
if torch.cuda.is_available():
    device_count = torch.cuda.device_count()
    current_device = torch.cuda.current_device()
    device_name = torch.cuda.get_device_name(current_device)
    
    print("üéâ CUDA –¥–æ—Å—Ç—É–ø–Ω–∞!")
    print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ GPU: {device_count}")
    print(f"–¢–µ–∫—É—â–µ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {current_device} ({device_name})")
else:
    print("‚ùóÔ∏è CUDA –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–∞. –ú–æ–¥–µ–ª—å –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–∞ CPU (–º–µ–¥–ª–µ–Ω–Ω–æ).")
    print("‚ùóÔ∏è –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤—ã –∑–∞–ø—É—Å—Ç–∏–ª–∏ –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä —Å —Ñ–ª–∞–≥–æ–º --gpus all.")

# –£—Å—Ç–∞–Ω–æ–≤–∏–º —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º '!' –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∫–æ–º–∞–Ω–¥—ã –≤ —Ç–µ—Ä–º–∏–Ω–∞–ª–µ –∏–∑ —è—á–µ–π–∫–∏ –Ω–æ—É—Ç–±—É–∫–∞
# –°–∫–∞—á–∏–≤–∞–µ–º —Ç–µ—Å—Ç–æ–≤–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ 'truck.jpg' –∏–∑ –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏—è SAM,
# —Ç–∞–∫ –∫–∞–∫ Pexels –±–ª–æ–∫–∏—Ä—É–µ—Ç –ø—Ä—è–º—ã–µ –∑–∞–ø—Ä–æ—Å—ã.
!mkdir -p /app/DATA
!wget -O /app/DATA/truck.jpg "https://raw.githubusercontent.com/facebookresearch/segment-anything/main/notebooks/images/truck.jpg"

print("‚úÖ –ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ 'truck.jpg' —É—Å–ø–µ—à–Ω–æ —Å–∫–∞—á–∞–Ω–æ –≤ –ø–∞–ø–∫—É /app/DATA/")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from segment_anything import sam_model_registry, SamAutomaticMaskGenerator

# --- 1. –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ø—É—Ç–∏ ---
# –ü—É—Ç—å –∫ —Å–∫–∞—á–∞–Ω–Ω–æ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é
image_path = "/app/DATA/truck.jpg"
# –ü—É—Ç—å –∫ –º–æ–¥–µ–ª–∏ –≤–Ω—É—Ç—Ä–∏ –∫–æ–Ω—Ç–µ–π–Ω–µ—Ä–∞
model_checkpoint_path = "/app/MODELS/sam_vit_h_4b8939.pth"
model_type = "vit_h"

# --- 2. –ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª—å ---
print("–ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏... –≠—Ç–æ –º–æ–∂–µ—Ç –∑–∞–Ω—è—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–µ–∫—É–Ω–¥.")
sam = sam_model_registry[model_type](checkpoint=model_checkpoint_path)
sam.to(device=device)
print("–ú–æ–¥–µ–ª—å —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –Ω–∞", device)

# --- 3. –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –º–∞—Å–æ–∫ ---
# SamAutomaticMaskGenerator –Ω–∞–π–¥–µ—Ç –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏
mask_generator = SamAutomaticMaskGenerator(sam)

# --- 4. –ó–∞–≥—Ä—É–∂–∞–µ–º –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ ---
# OpenCV —á–∏—Ç–∞–µ—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Ñ–æ—Ä–º–∞—Ç–µ BGR, –∞ –º–æ–¥–µ–ª—å –æ–∂–∏–¥–∞–µ—Ç RGB
image_bgr = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
print(f"–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∑–∞–≥—Ä—É–∂–µ–Ω–æ, –µ–≥–æ —Ä–∞–∑–º–µ—Ä: {image_rgb.shape}")

# --- 5. –ó–∞–ø—É—Å–∫–∞–µ–º —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—é ---
print("–ó–∞–ø—É—Å–∫ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –º–∞—Å–æ–∫... –≠—Ç–æ –º–æ–∂–µ—Ç –∑–∞–Ω—è—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –≤—Ä–µ–º—è –Ω–∞ GPU.")
masks = mask_generator.generate(image_rgb)
print(f"‚úÖ –ì–æ—Ç–æ–≤–æ! –ù–∞–π–¥–µ–Ω–æ {len(masks)} –º–∞—Å–æ–∫.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_anns_on_image(image_rgb, anns):
    """
    –°–æ–∑–¥–∞–µ—Ç –Ω–æ–≤–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ, –∫–æ–º–±–∏–Ω–∏—Ä—É—è –∏—Å—Ö–æ–¥–Ω–æ–µ —Å –º–∞—Å–∫–∞–º–∏.
    """
    if len(anns) == 0:
        return image_rgb

    # –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è, —á—Ç–æ–±—ã –Ω–µ –∏–∑–º–µ–Ω—è—Ç—å –æ—Ä–∏–≥–∏–Ω–∞–ª
    img = image_rgb.copy()
    
    # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ª–æ–π –¥–ª—è –º–∞—Å–æ–∫ —Å –∞–ª—å—Ñ–∞-–∫–∞–Ω–∞–ª–æ–º (–ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å—é)
    # –§–æ—Ä–º–∞—Ç: RGBA, –≥–¥–µ A - –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å
    overlay = np.zeros((img.shape[0], img.shape[1], 4), dtype=np.uint8)
    
    # –°–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å–∫–∏ –ø–æ –ø–ª–æ—â–∞–¥–∏, —á—Ç–æ–±—ã –±–æ–ª—å—à–∏–µ –Ω–µ –ø–µ—Ä–µ–∫—Ä—ã–≤–∞–ª–∏ –º–∞–ª–µ–Ω—å–∫–∏–µ
    sorted_anns = sorted(anns, key=(lambda x: x['area']), reverse=True)

    for ann in sorted_anns:
        m = ann['segmentation']
        # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–π —Ü–≤–µ—Ç –¥–ª—è –º–∞—Å–∫–∏
        color_mask = np.random.randint(0, 256, 3) # –°–ª—É—á–∞–π–Ω—ã–π RGB —Ü–≤–µ—Ç
        
        # –ü—Ä–∏–º–µ–Ω—è–µ–º —Ü–≤–µ—Ç –∫ –ø–∏–∫—Å–µ–ª—è–º, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –º–∞—Å–∫–µ
        # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å –≤ 90 (–∏–∑ 255)
        overlay[m, :3] = color_mask
        overlay[m, 3] = 90  # Alpha-–∫–∞–Ω–∞–ª (–ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç—å)

    # –ö–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –∏—Å—Ö–æ–¥–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤ —Ñ–æ—Ä–º–∞—Ç RGBA
    img_rgba = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)
    
    # –ù–∞–∫–ª–∞–¥—ã–≤–∞–µ–º —Å–ª–æ–π —Å –º–∞—Å–∫–∞–º–∏ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º cv2.addWeighted –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —Å–º–µ—à–∏–≤–∞–Ω–∏—è —Å —É—á–µ—Ç–æ–º –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç–∏
    # alpha=1.0 –¥–ª—è –∏—Å—Ö–æ–¥–Ω–æ–≥–æ, beta=1.0 –¥–ª—è –Ω–∞–ª–æ–∂–µ–Ω–∏—è, gamma=0
    final_image = cv2.addWeighted(img_rgba, 1.0, overlay, 1.0, 0)
    
    # –ö–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ RGB –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ matplotlib
    final_image_rgb = cv2.cvtColor(final_image, cv2.COLOR_RGBA2RGB)

    return final_image_rgb

# --- –°–æ–∑–¥–∞–µ–º –∏ –æ—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ ---
result_image = show_anns_on_image(image_rgb, masks)

plt.figure(figsize=(15, 10))
plt.imshow(result_image)
plt.axis('off')
plt.title("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–±–æ—Ç—ã Segment Anything –Ω–∞ 'truck.jpg'")
plt.show()
```

```python
# –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ —ç—Ç–∏ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã
import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- 1. –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –∫ –≤–∞—à–µ–π —Ñ–æ—Ç–æ–≥—Ä–∞—Ñ–∏–∏ ---
# –§–∞–π–ª –¥–æ–ª–∂–µ–Ω –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ –ø–∞–ø–∫–µ /app/CODE/
image_path = "/app/CODE/Vistino20241014_E57/1_normals.jpg"  # <-- –ò–ó–ú–ï–ù–ò–¢–ï –≠–¢–û –ò–ú–Ø –Ω–∞ –∏–º—è –≤–∞—à–µ–≥–æ —Ñ–∞–π–ª–∞!

# --- 2. –ó–∞–≥—Ä—É–∂–∞–µ–º –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –≤–∞—à–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ ---
# –ú–æ–¥–µ–ª—å –∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –º–∞—Å–æ–∫ —É–∂–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã –≤ –ø–∞–º—è—Ç—å, –∏—Ö –Ω–µ –Ω—É–∂–Ω–æ —Å–æ–∑–¥–∞–≤–∞—Ç—å –∑–∞–Ω–æ–≤–æ
try:
    image_bgr = cv2.imread(image_path)
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∑–∏–ª–æ—Å—å
    if image_bgr is None:
        raise FileNotFoundError(f"–ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç—å: {image_path}")
        
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
    print(f"–í–∞—à–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ '{image_path.split('/')[-1]}' —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ. –†–∞–∑–º–µ—Ä: {image_rgb.shape}")

    # --- 3. –ó–∞–ø—É—Å–∫–∞–µ–º —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—é ---
    print("–ó–∞–ø—É—Å–∫–∞—é —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—é –≤–∞—à–µ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è...")
    masks = mask_generator.generate(image_rgb)
    print(f"‚úÖ –ì–æ—Ç–æ–≤–æ! –ù–∞–π–¥–µ–Ω–æ {len(masks)} –º–∞—Å–æ–∫.")

    # --- 4. –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç ---
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ç—É –∂–µ —Ñ—É–Ω–∫—Ü–∏—é, —á—Ç–æ –∏ –¥–ª—è –≥—Ä—É–∑–æ–≤–∏–∫–∞
    result_image = show_anns_on_image(image_rgb, masks)

    plt.figure(figsize=(15, 10))
    plt.imshow(result_image)
    plt.axis('off')
    plt.title(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏ –¥–ª—è '{image_path.split('/')[-1]}'")
    plt.show()

except FileNotFoundError as e:
    print(e)
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –º–∞—Å–æ–∫
print(f"–í—Å–µ–≥–æ –º–∞—Å–æ–∫: {len(masks)}")

# –ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–∏–µ –∫–ª—é—á–∏ (–∫–∞–∫–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è) –µ—Å—Ç—å –¥–ª—è –∫–∞–∂–¥–æ–π –º–∞—Å–∫–∏
# –í–æ–∑—å–º–µ–º –ø–µ—Ä–≤—É—é –º–∞—Å–∫—É –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–∏–º–µ—Ä–∞ (–∏–Ω–¥–µ–∫—Å 0)
if len(masks) > 0:
    print("\n–ö–ª—é—á–∏ –¥–ª—è –æ–¥–Ω–æ–π –º–∞—Å–∫–∏:")
    print(list(masks[0].keys()))
```

```python
# --- –í–´–ë–ï–†–ò–¢–ï –ò–ù–î–ï–ö–° –ú–ê–°–ö–ò –î–õ–Ø –ò–°–°–õ–ï–î–û–í–ê–ù–ò–Ø ---
mask_index = 3  # –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –º–µ–Ω—è—Ç—å —ç—Ç–æ —á–∏—Å–ª–æ –æ—Ç 0 –¥–æ 34

# –ü–æ–ª—É—á–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–∞—Å–∫–∏
single_mask_data = masks[mask_index]

# –ò–∑–≤–ª–µ–∫–∞–µ–º —Å–∞–º—É –º–∞—Å–∫—É (2D –º–∞—Å—Å–∏–≤ True/False)
mask = single_mask_data['segmentation']

# –ò–∑–≤–ª–µ–∫–∞–µ–º –µ–µ –ø–ª–æ—â–∞–¥—å –∏ —Ä–∞–º–∫—É
area = single_mask_data['area']
bbox = single_mask_data['bbox']

print(f"–ò—Å—Å–ª–µ–¥—É–µ–º –º–∞—Å–∫—É —Å –∏–Ω–¥–µ–∫—Å–æ–º {mask_index}")
print(f"–ï–µ –ø–ª–æ—â–∞–¥—å: {area} –ø–∏–∫—Å–µ–ª–µ–π")
print(f"–ï–µ —Ä–∞–º–∫–∞ (bbox): {bbox}")

# –†–∏—Å—É–µ–º —Ç–æ–ª—å–∫–æ —ç—Ç—É –º–∞—Å–∫—É
plt.figure(figsize=(10, 10))
plt.imshow(mask, cmap='gray') # cmap='gray' –ø–æ–∫–∞–∂–µ—Ç –µ–µ –≤ —á–µ—Ä–Ω–æ-–±–µ–ª–æ–º –≤–∏–¥–µ
plt.axis('off')
plt.title(f"–ú–∞—Å–∫–∞ ‚Ññ{mask_index}")
plt.show()

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ç—É –∂–µ –≤—ã–±—Ä–∞–Ω–Ω—É—é –º–∞—Å–∫—É –∏–∑ —è—á–µ–π–∫–∏ –≤—ã—à–µ
# (—É–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ mask_index, mask, image_rgb —Å—É—â–µ—Å—Ç–≤—É—é—Ç)

# –°–æ–∑–¥–∞–µ–º —Ü–≤–µ—Ç–Ω—É—é –≤–µ—Ä—Å–∏—é –º–∞—Å–∫–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, —è—Ä–∫–æ-–∫—Ä–∞—Å–Ω—É—é)
highlight_color = [255, 0, 0] # –ö—Ä–∞—Å–Ω—ã–π
highlight_mask = np.zeros_like(image_rgb)
highlight_mask[mask] = highlight_color

# –°–º–µ—à–∏–≤–∞–µ–º –∏—Å—Ö–æ–¥–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Å –Ω–∞—à–µ–π –ø–æ–¥—Å–≤–µ—Ç–∫–æ–π
# 0.6 –∏ 0.4 - –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–æ–∑—Ä–∞—á–Ω–æ—Å—Ç–∏
highlighted_image = cv2.addWeighted(image_rgb, 0.6, highlight_mask, 0.4, 0)

plt.figure(figsize=(15, 10))
plt.imshow(highlighted_image)
plt.axis('off')
plt.title(f"–ü–æ–¥—Å–≤–µ—Ç–∫–∞ –º–∞—Å–∫–∏ ‚Ññ{mask_index}")
plt.show()
```
